```python
# ==============================================
# 1. FIRST, INSTALL REQUIRED LIBRARIES
# ==============================================
!pip install scikit-fuzzy numpy matplotlib -q

# ==============================================
# 2. NOW IMPORT EVERYTHING
# ==============================================
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

print("‚úÖ Libraries imported successfully!")

# ==============================================
# 3. CREATE ALL VARIABLES IN ORDER
# ==============================================

# --- CREATE OUTPUT VARIABLE FIRST ---
print("üîß Creating output variable (risk)...")
risk = ctrl.Consequent(np.arange(0, 101, 1), 'risk')
risk['low'] = fuzz.trimf(risk.universe, [0, 0, 30])
risk['medium'] = fuzz.trimf(risk.universe, [20, 50, 80])
risk['high'] = fuzz.trimf(risk.universe, [70, 90, 100])
print("‚úÖ Risk variable created!")

# --- CREATE INPUT VARIABLES ---
print("üîß Creating input variables...")

# Temperature
temperature = ctrl.Antecedent(np.arange(35, 41, 0.1), 'temperature')
temperature['hypothermic'] = fuzz.trimf(temperature.universe, [35, 35, 36])
temperature['normal'] = fuzz.trimf(temperature.universe, [35.5, 36.5, 37.5])
temperature['fever'] = fuzz.trimf(temperature.universe, [37, 37.8, 38.5])
temperature['high_fever'] = fuzz.trimf(temperature.universe, [38, 39, 41])

# Duration
duration = ctrl.Antecedent(np.arange(0, 15, 0.1), 'duration')
duration['acute'] = fuzz.trimf(duration.universe, [0, 0, 3])
duration['subacute'] = fuzz.trimf(duration.universe, [2, 5, 8])
duration['chronic'] = fuzz.trimf(duration.universe, [7, 11, 15])

# Age
age = ctrl.Antecedent(np.arange(0, 100, 1), 'age')
age['child'] = fuzz.trimf(age.universe, [0, 0, 18])
age['adult'] = fuzz.trimf(age.universe, [15, 40, 65])
age['elderly'] = fuzz.trimf(age.universe, [60, 80, 100])

print("‚úÖ All input variables created!")

# ==============================================
# 4. CREATE FUZZY RULES (NOW RISK IS DEFINED)
# ==============================================

print("\nüîß Creating fuzzy rules...")

try:
    rule1 = ctrl.Rule(temperature['normal'] & duration['acute'], risk['low'])
    rule2 = ctrl.Rule(temperature['fever'] & duration['acute'], risk['medium'])
    rule3 = ctrl.Rule(temperature['high_fever'] & duration['acute'], risk['high'])
    rule4 = ctrl.Rule(temperature['fever'] & duration['subacute'], risk['high'])
    rule5 = ctrl.Rule(temperature['normal'] & duration['chronic'] & age['elderly'], risk['medium'])
    rule6 = ctrl.Rule(temperature['fever'] & duration['chronic'] & age['elderly'], risk['high'])

    print(f"‚úÖ Created {6} fuzzy rules successfully!")

except Exception as e:
    print(f"‚ùå Error creating rules: {e}")
    print("Make sure 'risk' variable is created before rules!")
    exit()

# ==============================================
# 5. CREATE CONTROL SYSTEM
# ==============================================

print("\nüîß Creating control system...")

try:
    risk_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6])
    risk_simulation = ctrl.ControlSystemSimulation(risk_ctrl)
    print("‚úÖ Control system created!")

except Exception as e:
    print(f"‚ùå Error creating control system: {e}")
    exit()

# ==============================================
# 6. TEST THE SYSTEM WITH SAMPLE CASES
# ==============================================

print("\n" + "="*50)
print("üß™ TESTING THE SYSTEM")
print("="*50)

# Define test cases
test_cases = [
    {"name": "Adult with acute fever", "temp": 38.2, "days": 2, "age": 35},
    {"name": "Elderly with chronic fever", "temp": 37.9, "days": 10, "age": 72},
    {"name": "Child with normal temp", "temp": 36.8, "days": 1, "age": 8},
    {"name": "Adult with high fever", "temp": 39.5, "days": 3, "age": 45},
    {"name": "Hypothermic elderly", "temp": 35.5, "days": 5, "age": 70}
]

results = []

for case in test_cases:
    print(f"\nüìã {case['name']}:")
    print(f"   Temperature: {case['temp']}¬∞C")
    print(f"   Duration: {case['days']} days")
    print(f"   Age: {case['age']} years")

    try:
        risk_simulation.input['temperature'] = case['temp']
        risk_simulation.input['duration'] = case['days']
        risk_simulation.input['age'] = case['age']

        risk_simulation.compute()
        risk_value = risk_simulation.output['risk']

        # Categorize
        if risk_value < 30:
            category = "LOW"
            action = "‚úÖ Home care"
            color = "üü¢"
        elif risk_value < 70:
            category = "MEDIUM"
            action = "‚ö†Ô∏è GP consultation"
            color = "üü°"
        else:
            category = "HIGH"
            action = "üö® Urgent medical attention"
            color = "üî¥"

        print(f"   {color} Infection Risk: {risk_value:.1f}% ({category})")
        print(f"   Recommendation: {action}")

        results.append({
            "case": case['name'],
            "risk": risk_value,
            "category": category
        })

    except Exception as e:
        print(f"   ‚ùå Error: {e}")

# ==============================================
# 7. CREATE VISUALIZATIONS
# ==============================================

print("\nüìä Creating visualizations...")

try:
    # Create figure with 4 subplots
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    # Temperature
    temperature.view(ax=axes[0, 0])
    axes[0, 0].set_title('Temperature (¬∞C)', fontsize=12, fontweight='bold')
    axes[0, 0].set_xlabel('Temperature')
    axes[0, 0].set_ylabel('Membership')

    # Duration
    duration.view(ax=axes[0, 1])
    axes[0, 1].set_title('Symptom Duration (days)', fontsize=12, fontweight='bold')
    axes[0, 1].set_xlabel('Duration')
    axes[0, 1].set_ylabel('Membership')

    # Age
    age.view(ax=axes[1, 0])
    axes[1, 0].set_title('Age (years)', fontsize=12, fontweight='bold')
    axes[1, 0].set_xlabel('Age')
    axes[1, 0].set_ylabel('Membership')

    # Risk
    risk.view(ax=axes[1, 1])
    axes[1, 1].set_title('Infection Risk (%)', fontsize=12, fontweight='bold')
    axes[1, 1].set_xlabel('Risk')
    axes[1, 1].set_ylabel('Membership')

    plt.suptitle('Fuzzy Logic Membership Functions', fontsize=16, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.savefig('fuzzy_membership_functions.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("‚úÖ Visualizations saved as 'fuzzy_membership_functions.png'")

except Exception as e:
    print(f"‚ùå Error creating visualizations: {e}")

# ==============================================
# 8. CREATE SIMPLE INTERACTIVE FUNCTION
# ==============================================

def quick_assess():
    """Simple interactive assessment"""
    print("\n" + "="*50)
    print("üë®‚Äç‚öïÔ∏è QUICK PATIENT ASSESSMENT")
    print("="*50)

    try:
        # Test values (you can change these)
        temp = 38.5
        days = 3
        age_years = 40

        print(f"Using test values: Temp={temp}¬∞C, Duration={days} days, Age={age_years}")

        risk_simulation.input['temperature'] = temp
        risk_simulation.input['duration'] = days
        risk_simulation.input['age'] = age_years
        risk_simulation.compute()
        risk_value = risk_simulation.output['risk']

        print(f"\nüìã ASSESSMENT RESULTS:")
        print(f"   Temperature: {temp}¬∞C")
        print(f"   Symptom Duration: {days} days")
        print(f"   Age: {age_years} years")
        print(f"   Infection Risk: {risk_value:.1f}%")

        if risk_value < 30:
            print("   üü¢ LOW RISK: Home care advised")
        elif risk_value < 70:
            print("   üü° MEDIUM RISK: GP consultation recommended")
        else:
            print("   üî¥ HIGH RISK: Urgent medical attention needed")

        return risk_value

    except Exception as e:
        print(f"‚ùå Error: {e}")
        return None

# Run quick assessment
quick_assess()

# ==============================================
# 9. SUMMARY FOR YOUR ASSIGNMENT
# ==============================================

print("\n" + "="*60)
print("üìù SUMMARY FOR YOUR ASSIGNMENT REPORT")
print("="*60)
print("\nInclude these in your Task 1 submission:")
print("1. ‚úÖ Fuzzy Logic Concepts:")
print("   - Variables: Temperature, Duration, Age, Risk")
print("   - Membership functions (see saved plot)")
print("   - 6 fuzzy rules for clinical decision-making")
print("   - Mamdani inference system")
print("\n2. ‚úÖ Test Results:")
for result in results:
    print(f"   - {result['case']}: {result['risk']:.1f}% ({result['category']})")
print("\n3. ‚úÖ How This Demonstrates Fuzzy Logic:")
print("   - Handles uncertainty (e.g., 'fever' is not binary)")
print("   - Uses linguistic variables (low/medium/high)")
print("   - Provides explainable decisions through rules")
print("   - Shows real-world healthcare application")
print("\n4. ‚úÖ Code Output:")
print("   - Screenshot of test cases above")
print("   - Saved visualization: fuzzy_membership_functions.png")
print("\n5. ‚úÖ Connection to Your Case Studies:")
print("   - Like OCT classification: handles uncertainty in medical data")
print("   - Like patient trust systems: provides explainable decisions")
print("   - Like literature review: uses rule-based reasoning")
print("="*60)
```
